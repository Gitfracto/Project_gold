```python
import pandas as pd
import re
#import the needed libraries
```

```python
df = pd.read_csv("XAUUSD.csv")
#import the file we will use.
#got it from duka.com
```

```python
df
#see our dataframe
```

```python
df = df.drop(["Open","High","Low",'Volume'],axis=1)
#drop the columns that we dont need
```

```python
df.rename(columns = {'Local time':'time'}, inplace = True)
df.rename(columns = {'Close':'gold_price'}, inplace = True)

#rename the columns
```

```python
def clean_gmt(x):
    return x.replace("GMT+0200","").replace("GMT+0100","").replace("23:00:00.000","")
#function to take out the GMT part of the date
```

```python
df["time"] = df.time.apply(clean_gmt)
#apply the function
```

```python
def cleaning_day(x):
    y = x.split('.')
    return int(y[0])
#splitting the day form the date
```

```python
def cleaning_month(x):
    y = x.split('.')
    return int(y[1])
#splitting the month form the date
```

```python
def cleaning_year(x):
    return int(re.findall(r'\b\d{4}\b',x)[0])
#splitting the year form the date. This one had to be done like this because it 
#would come like "yyyy.0" and not "yyyy" as expected.
```

```python
df["day"] = df.time.apply(cleaning_day)
df["month"] = df.time.apply(cleaning_month)
df["year"] = df.time.apply(cleaning_year)
#apply all functions
```

```python
df.drop(df[df.year > 2021].index, inplace=True)
#drop any price after 2021
```

```python
df.drop(df[df.year < 2020].index, inplace=True)
#drop any price before 2020
```

```python
df = df.drop(["time"],axis=1)
#drop time because we already divided it in day,month,year
```

```python
df = df.iloc[:, [3,2,1,0]]
#organize the df columns in the order we want
```

```python
df.to_csv('gold_clean.csv',encoding='utf-8-sig',index=False)
#export the cleaned df to gold_clean
```
